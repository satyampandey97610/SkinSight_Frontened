```python
 
import os
from PIL import Image

TARGET_SIZE = (224, 224)
SRC_FOLDER = "Dataset_Minor_Skin"
DST_FOLDER = "Dataset_Minor_Skin_Resized"

os.makedirs(DST_FOLDER, exist_ok=True)

# Loop through each class folder
for class_name in os.listdir(SRC_FOLDER):
    class_path = os.path.join(SRC_FOLDER, class_name)
    save_path = os.path.join(DST_FOLDER, class_name)
    os.makedirs(save_path, exist_ok=True)

    if not os.path.isdir(class_path):
        continue

    print(f"Processing: {class_name}")

    for img_name in os.listdir(class_path):
        try:
            img_path = os.path.join(class_path, img_name)
            img = Image.open(img_path).convert("RGB")
            img = img.resize(TARGET_SIZE)
            img.save(os.path.join(save_path, img_name))
        except Exception as e:
            print(f"‚ùå Failed on {img_name}: {e}")
```
